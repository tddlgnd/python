```python
# 3dgraph.py
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 

# 3 (axis) . 
axis = plt.axes(projection='3d')
#plt.show()

# 3    . 
Z = np.linspace(0, 1, 100)
X = Z * np.sin(30 * Z)
Y = Z * np.cos(30 * Z)

# 3  . 
axis.plot3D(X, Y, Z)
```

```python
# gaussian.py
import numpy as np
import matplotlib.pyplot as plt

m, sigma = 10, 2
Y1 = np.random.randn(10000)
Y2 = m+sigma*np.random.randn(10000)

plt.figure(figsize=(10,6))		#   
plt.hist(Y1, bins=20)
plt.hist(Y2, bins=20)
plt.show()
```

```python
# draw_sine.py
import matplotlib.pyplot as plt
import numpy as np

# -2 +2 100   . 
X = np.linspace(-2 * np.pi, 2 * np.pi, 100)

#   sin()  . 
Y1 = np.sin(X)
Y2 = 3 * np.sin(X)

plt.plot(X, Y1, X, Y2)
plt.show()
```

```python
# calc_mse.py
import numpy as np

ypred = np.array([1, 0, 0, 0, 0])
y = np.array([0, 1, 0, 0, 0])
n = 5
MSE = (1/n) * np.sum(np.square(ypred-y))
print(MSE)
```
